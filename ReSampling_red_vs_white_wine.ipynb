```python
# Import dependencies 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy
```

```python
#load/read the data with pandas
red_wine = pd.read_csv("winequality-red (1).csv")
white_wine = pd.read_csv("winequality-white.csv")
```

```python
#put the data into dataframes
red_db = pd.DataFrame(red_wine)
white_db = pd.DataFrame(white_wine)
```

```python
white_sample = white_db.sample(n = 666 , random_state=numpy.random.RandomState())
red_sample = red_db.sample(n = 666 , random_state=numpy.random.RandomState())

ALL_data = pd.concat([red_sample, white_sample],axis=0)
```

```python
red_correlation = red_db.corr()
# display(correlation)
plt.figure(figsize=(14, 12))
heatmap_red = sns.heatmap(red_correlation, annot=True, linewidths=0, vmin=-1, cmap="RdBu_r")
red_correlation.to_csv('to_csv/red_correlation.csv')
```

```python
white_correlation = white_db.corr()
# display(correlation)
plt.figure(figsize=(14, 12))
heatmap_white = sns.heatmap(white_correlation, annot=True, linewidths=0, vmin=-1, cmap="RdBu_r")
white_correlation.to_csv('to_csv/white_correlation.csv')
```

```python
all_correlation = ALL_data.corr()
# display(correlation)
plt.figure(figsize=(14, 12))
heatmap_all = sns.heatmap(all_correlation, annot=True, linewidths=0, vmin=-1, cmap="RdBu_r")
all_correlation.to_csv('to_csv/all_correlation.csv')
```

# White Wine Data Evaluation

```python
#scope the white wine data
white_db.describe()
```

```python
#any missing data?
white_db.isnull().any()
```

```python
#sort white_db
white_db.sort_values(by=['quality'], ascending=False)
```

```python
# Classify the Quality of wines
quality = white_db["quality"].values
category = []
for num in quality:
    if num<5:
        category.append("Poor")
    elif num>6:
        category.append("Good")
    else:
        category.append("Avg")
```

```python
#Create new data
category = pd.DataFrame(data=category, columns=["category"])
WW_data = pd.concat([white_db,category],axis=1)

#WW_data.to_csv("Tableau_White_Wine.csv", sep = ",")

WW_data.drop(columns="quality",axis=1,inplace=True)
# Show new Data
WW_data.head()
```

```python
quality_good = WW_data.loc[WW_data['category'] == 'Good']
quality_avg = WW_data.loc[WW_data['category'] == 'Avg']
quality_poor = WW_data.loc[WW_data['category'] == 'Poor']

quality_poor.head(5)
```

```python

good_sample = quality_good.sample(n = 183 , random_state=numpy.random.RandomState())
poor_sample = quality_poor.sample(n = 183 , random_state=numpy.random.RandomState())
avg_sample = quality_avg.sample(n = 183 , random_state=numpy.random.RandomState())
avg_sample.head(5)
```

```python
WW_data = []
WW_data = pd.concat([good_sample,poor_sample,avg_sample],axis=0)
WW_data
```

```python
# Visualize Category column 
plt.figure(figsize=(10,6))
sns.countplot(WW_data["category"],palette="Greens")
WW_data["category"].value_counts()
```

```python
# Visualize Alcohol and Quality
plt.figure(figsize=(12,6))
sns.barplot(x=white_db["quality"],y=white_db["alcohol"],palette="Greens")
```

# Building Machine Learning

## Testing Red Wine

```python
# Grabbing the array of numbers and 
X= RW_data.iloc[:,:-1].values
y= RW_data.iloc[:,-1].values
```

```python
X
```

```python
y
```

```python
# setting quality values into numeric values (0: avg, 1: good, 2:poor)

from sklearn.preprocessing import LabelEncoder
labelencoder_y =LabelEncoder()
y= labelencoder_y.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2,random_state=50)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train,y_train)
pred_svc =svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score
print(classification_report(y_test,pred_svc))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=250)
rfc.fit(X_train, y_train)
pred_rfc = rfc.predict(X_test)
print(classification_report(y_test, pred_rfc))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
pred_knn=knn.predict(X_test)
print(classification_report(y_test, pred_knn))
```

```python
red_wine_conclusion = pd.DataFrame({'models': ["SVC","Random Forest","KNN"],
                           'accuracies': [accuracy_score(y_test,pred_svc),accuracy_score(y_test,pred_rfc),accuracy_score(y_test,pred_knn)]})
red_wine_conclusion
```
